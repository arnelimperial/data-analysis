```python
plt.show()
```

```python
%matplotlib inline
```

```python
sns.set_style('darkgrid')
```

```python
df1 = pd.read_csv('df1',index_col=0)
```

```python
df1.head()
```

```python
type(df1)
```

```python
df2 = pd.read_csv('df2')
```

```python
df2
```

```python
#Matplotlib is applicable to all plots
# Histogram for df1
df1.hist(bins=30)
plt.tight_layout()
```

```python
df1['A'].hist(bins=30)
plt.show()
```

```python
df1['A'].plot(kind='hist',bins=20)
plt.show()
```

```python
# Advisable
df1['A'].plot.hist()
```

```python
df1.plot.hist(bins=30,color='purple')
```

```python
df2
```

```python
# Area Plot
df2.plot.area(alpha=0.4)
```

```python
# Bar Plot (Categorical)
df2.plot.bar()
```

```python
df2.plot.bar(stacked=True)
```

```python
# Line plot
ind = df1.index.to_series()
ind
```

```python
# x=df1.index,y='B'
df1.plot.line(y='B',figsize=(12,3),lw=1)
```

```python
# Scatter Plot
df1.plot.scatter(x='A',y='B')
```

```python
# 3 Dimensional Plot with scatter plot of Column D
df1.plot.scatter(x='A',y='B',c='D',cmap='gnuplot_r')
```

```python
df2.plot.scatter(x='a',y='b',s=df2['d']*100)
```

```python
# Box plot
df2.plot.box()
```

```python
df1.plot.box()
```

```python
df = pd.DataFrame(np.random.randn(1000,2),columns=['A','B'])
df.head()
```

```python
# Bi Variate hexplot
df.plot.hexbin(x='A',y='B',gridsize=25,cmap='Spectral')
# Darker the color is where data is concentrated at
```

```python
# Kernel Density Estimation
df2['a'].plot.kde()
```

```python
df2.plot.density()
```
