# Seaborn: Distribution Plot

```python
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#Load built-in dataset called tips
tips = sns.load_dataset('tips')
plt.show()
```

```python
tips.head()
```

```python
#distplot with kde
sns.distplot(tips['total_bill'])
```

```python
#distplot w/o kde
sns.distplot(tips['total_bill'], kde=False,bins=20)
```

```python
# Combining & compairing two distribution plot; with 2 variable
sns.jointplot(x='total_bill',y='tip',data=tips)

plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
j = sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
j.annotate(stats.pearsonr)
#plt.show()
```

```python
j = sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
j.annotate(stats.pearsonr)
```

```python
# Pair Plot
sns.pairplot(tips, hue='sex',palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])
```

```python
# KDE kernel density estimation plot
sns.kdeplot(tips['total_bill'])
```
