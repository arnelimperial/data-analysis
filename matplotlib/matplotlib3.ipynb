```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
%matplotlib inline
```

```python
# Sample Data
# Create a dataframe with an x column containing values from -10 to 10
d = {'x': range(-10,11)}
df = pd.DataFrame({'x': range(-10,11)})
# Add a y column by applying the slope-intercept equation to x
df['y'] = 2.0**df['x']
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
# RGB code or color string e.g.:red
# linewidth can be written as lw
#alpha for line transparency
#linestyle can be written as ls
#markers for dots of x vs y
ax.plot(df.x, df.y, 
        color='#FF8C08',
        lw=1,alpha=0.8,ls='--',
        marker='o', 
        markersize=10, 
        markerfacecolor='red',
        markeredgewidth=1,
        markeredgecolor='blue',
       ) 

plt.show()
```

```python
fig1 = plt.figure()

ax = fig1.add_axes([0,0,1,1])
# RGB code or color string e.g.:red
# linewidth can be written as lw
#alpha for line transparency
#linestyle can be written as ls
#markers for dots of x vs y
ax.plot(df.x, df.y,color='#FF8C08',lw=2,ls='--') 

#[lowerbound,upperbound]
ax.set_xlim([0,1])
ax.set_ylim([0,3])


plt.show()
```
