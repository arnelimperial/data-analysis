# OO Method

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
#fig = plt.figure()
#axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])# list from 0 - 1[left, bottom, width, height]
#axes.plot(x,y)
#axes.set_xlabel('X')
#axes.set_ylabel('Y')
#axes.set_title('Sample Plot')


#plt.show()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
```

```python
fig, axes = plt.subplots(nrows=3,ncols=3)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
#axes obj is iterable

for curr_ax in axes:
    curr_ax.plot(x,y)
```

```python
#Doing index call
#setxlabel and setylabel also applied
fig, axes = plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y)
axes[0].set_title('First Plot')
axes[1].plot(y,x)
axes[1].set_title('Second Plot')


plt.tight_layout()
```

# Figure Size and DPI

```python
#figsize tuple is width and height of the figure in inches; dpi dot per inch
fig1 = plt.figure(figsize=(3,2))
ax = fig.add_axes([0,0,1,1])
plt.plot(x,y)
```

```python
#With subplots
fig2, ax = plt.subplots(figsize=(8,2))
ax.plot(x,y)
```

```python
#Index call
fig3, ax = plt.subplots(nrows=2, ncols=1,figsize=(8,2))
ax[0].plot(x,y)
ax[1].plot(y,x)

plt.tight_layout()
```
