# Matplotlib: Concepts

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
%matplotlib inline
```

```python
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
# Functional Method
plt.plot(x, y)

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample Plot')

plt.show() # Show to display plot not in jupyter notebook
```

```python
#Subplot [#rows, #cols, plot #]
plt.subplot(1,2,1)
plt.plot(x,y, 'r')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample Plot 1')

#########

plt.subplot(1,2,2)
plt.plot(y,x, 'b')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Sample Plot 2')



plt.show()
```

```python
# Object Oriented Method
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])# list from 0 - 1[left, bottom, width, height]
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Sample Plot')
plt.show()
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig1.add_axes([0.2, 0.5, 0.4, 0.3])


axes1.plot(x,y,'r')
axes2.plot(y,x,'m')
```

```python
fig, axes = plt.subplots(nrows=1,ncols=2)
axes.plot(x,y)
#plt.plot(x,y)
```
